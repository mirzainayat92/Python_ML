```python
import numpy as np
houses = []
houses = np.genfromtxt("Housing Price data set.csv",delimiter=",",skip_header=1,skip_footer = 146,usecols = (1,2,3,4))
houses
#1st column is price 2nd,3rd and 4th are input data 
#Taking first 400 samples as training data
```

```python
houses.shape
```

```python
house_Y = houses[:,[0]]
house_Y
```

```python
house_X = houses[:,1:4]
house_X
```

```python
X_transpose = house_X.transpose()
X_transpose
```

```python
X_T_X = np.dot(X_transpose,house_X)
X_T_X
```

```python
X_T_Y = np.dot(X_transpose,house_Y)
X_T_Y
```

```python
X_T_X_inverse = np.linalg.inv(X_T_X)
X_T_X_inverse
```

```python
theta = np.dot(X_T_X_inverse,X_T_Y)
theta
```

```python
#testing on data
test_houses = np.genfromtxt("Housing Price data set.csv",delimiter=",",skip_header=401,usecols = (1,2,3,4))
X_test = test_houses[:,1:4]
Y_test = test_houses[:,[0]]
Y_test
```

```python
Y_predict = np.dot(X_test,theta)
Y_predict
```

```python
#Root Mean Square Error
error_0_rms = np.mean([(Y_predict[i] - Y_test[i]) ** 2 for i in range(146)]) ** 0.5
error_0_rms 
```

```python
#Mean Absolute Percent Error
MAPE_0 = np.mean([abs((Y_predict[i] - Y_test[i])/Y_test[i]) for i in range(146)])*100
MAPE_0
```

```python
#Now coming to Normal Equation with Regularization
#here theta = (X_T_X + lambda(.)L).(X_T).(Y)
lambda_ = 1500
L = np.zeros(shape = (3,3))
L[1,1] = 1
L[2,2] = 1
L = lambda_ * L
L
```

```python
theta_1 = np.dot(np.linalg.inv(X_T_X + L),X_T_Y)
theta_1
```

```python
Y_predict_1 = np.dot(X_test,theta_1)
Y_predict_1
```

```python
#Root Mean Square Error
error_1 = np.mean([(Y_predict_1[i] - Y_test[i]) ** 2 for i in range(146)]) ** 0.5
error_1 
```

```python
#Mean Absolute Percent Error
MAPE_1 = np.mean([abs((Y_predict_1[i] - Y_test[i])/Y_test[i]) for i in range(146)])*100
MAPE_1
```

```python
''' After Comparing RMS Error value and MAP Error of above two models(i.e. Normal Equation with and 
    without Reguralisation with lambda value 1500),we can say that here Simple Normal Equation Model 
    is working better than the latter one.[We can change the value of lambda to view better results ]'''  
```
